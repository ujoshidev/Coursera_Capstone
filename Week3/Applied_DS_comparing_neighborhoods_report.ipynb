# Clustering and Comparing neighborhoods of New York and Toronto

## Introduction

In this project, we will analyze, cluster and compare the neighborhoods of two major
cities in the world: New York City which is located in United States of America and Toronto which
is located in Canada. <br>We will investigate what kinds of businesses are common in both cities
and compare the businesses. Also, we will investigate on what
kinds of businesses are not common in both cities.<br>


Doing this project will enable us to get a better understanding of similarities and differences
between these cities which will benefit all business people and give answers to the general questions like:<br> 
* what types of businesses are more likely to thrive in both cities?<br> 
* what are the neighborhoods that are suitable for each type of
business?<br>
* what types of businesses are not very desirable in each city?<br/>

This allows business people to take better and more effective decisions regarding where to open their businesses.

## Data Acquisition and Preparation

In this section, the processes of acquiring, cleaning, and preparing each dataset used in this project
for next stages will be specified.<br>
To be able to do this project, two types of data are needed:<br>
* **Neighborhood Data:** Datasets that lists the names of the neighborhoods of NYC and Toronto
and their latitude and longitude coordinates. We have some of this data provided by the
coordinators of "IBM Data Science Professional Certificate" and we also need to scrape some
data from the internet (Wikipedia).<br>
* **Venues data:** Data that describes the top 100 venues (restaurants, cafes, parks, museums, etc.)
in each neighborhood of the two cities. The data should list the venues of each neighborhood
with their categories. 
This data will be retrieved from Foursquare which is one of the world largest sources of
location and venue data. <br><br>

1. **Neighborhood Data**<br>
For each city, data that describes the names of its neighborhoods and their coordinates is needed.<br>
  1.1. **New York City**<br>
  A  dataset that specifies the neighborhood data for New York City was provided by the organizers
of “Applied Data Science Capstone” course which is provided by IBM. The dataset is originally a
JSON file that specifies the name of each neighborhood, its coordinates—latitude and longitude,
its borough, and other data too. <br>Below figure showing a part of this JSON file.<br>
image.png
<br> Figure: Part of the JSON file that describes NYC neighborhood<br>
To be able to use the data of this JSON file in the later parts of this project, it should be stored in
a Pandas dataframe. And then we will draw a map of NYC and its neighborhoods using
Folium Python package.

  1.2. **Toronto**<br>
  For Toronto, no dataset is provided in the course.<br>
  There is a Wikipedia page titled “List of postal codes of Canada: M”. This page lists the postal
codes in Canada that start with the letter M which are the postal codes of Toronto city.<br> It lists
the postal codes with the neighborhood and borough name associated with each postal code. <br>To
download this web page and extract the relevant data from it, Pandas read_html() functions can
be used. It reads HTML tables on a web page in a list of dataframes.<br>
Then we use the extracted information to chart a map of Toronto and its neighborhoods using Folium Python package.<br><br>
2. **Venues:** <br>
For each city, data that describes the venues of its neighborhoods and the categories of these venues
is needed. <br>Venues data will be retrieved from Foursquare which is a popular source of location and
venue data. Foursquare API service will be utilized to access and download venues data.<br>
To retrieve data from Foursquare using their API, a URL should be prepared and used to request data related a specific location.<br> An example URL is the following:<br>
  *https://api.foursquare.com/v2/venues/search?
&client_id=1234&client_secret=1234&v=20180605&
ll=40.89470517661,-73.84720052054902&radius=500&limit=100* <br><br>
where 
  * *search* indicates the API endpoint used
  * *client_id* and client_secret are credentials used
to access the API service and are obtained when registering a Foursquare developer account
  * *v*
indicates the API version to use
  * *ll* indicates the latitude and longitude of the desired location
  * *radius* is the maximum distance in meters between the specified location and the retrieved venues
  * *limit* is used to limit the number of returned results if necessary<br>
